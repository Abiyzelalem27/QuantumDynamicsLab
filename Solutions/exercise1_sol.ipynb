## Conclusion

The analytical eigenstates of the quantum harmonic oscillator were plotted on a spatial grid to study their mathematical structure and physical behavior. The lowest-energy states are smooth and clearly separated, while higher-energy states become increasingly oscillatory and contain more nodes. This confirms the quantum mechanical rule that the \(n\)-th eigenstate has \(n\) nodes.

The corresponding eigenenergies are equally spaced, following the relation:

\[
E_n = n + \frac{1}{2}
\]

which is a fundamental property of the harmonic oscillator. As more eigenstates are included, the spectrum becomes denser and visually more complex, demonstrating the infinite set of allowed quantized states.

The plots also illustrate the correspondence principle: for large quantum numbers, the probability densities spread over the classically allowed region and begin to resemble classical motion. Therefore, the harmonic oscillator provides an important model for understanding quantization, wavefunctions, and the connection between quantum and classical physics.

```python
# define spatial grid
L = 20
Ns = [100, 200, 401, 800]   # different grid sizes
results = []

for N in Ns:
    # Grid
    x = np.linspace(-L/2, L/2, N)
    dx = x[1] - x[0]
    
    # The Homiltonian 
    T = cqd.hamiltonians.H_kinetic(x)
    V = cqd.hamiltonians.HO_potential(x) 
    H = T + V
    
    # eigenenergies and eigenfunction
    E, psi = eigh(H)

    # store lowest 6 eigenvalues
    results.append(E[:6])
    print(f"N = {N}")
    print("E0–E5 =", E[:6])
    print()
# Compare convergence
print("Comparison table=\n", np.array(results))
```

The analytical ground state energy of the harmonic oscillator is 0.5.

When using a small grid (e.g., (N = 100)), the numerical ground state energy shows noticeable error.
As the number of grid points increases ((N = 200, 401, 800)), the numerical result converges toward the analytical value.
This occurs because a finer grid improves the finite-difference approximation of the kinetic energy operator, leading to more accurate eigenvalues.

The difference between the probability densities is small, confirming that the numerical method accurately reproduces the analytical harmonic oscillator solution.

```python
# define spatial grid
L = 10
n = 401

# The Homiltonian 
T = cqd.hamiltonians.H_kinetic(x)
V = cqd.hamiltonians.HO_potential(x) 
H = T + V

# diagonalize: here the whole magic is happening
evals, evecs = LA.eigh(H)

#check that the state is normalized (if we want the integral to be normalized we have to divide by dx)
n_plot = 0
LA.norm(evecs[:,n_plot])

# check convergence: relative error of eigenvalues
evals_exact = cqd.hamiltonians.HO_eigenenergies_exact(np.arange(evals.size))
(evals - evals_exact)[0:20]
n_plots= 1

# compare exact to numerical
plt.figure()
plt.plot(x, np.abs(cqd.hamiltonians.HO_eigenstates_exact(n_plot, x)) ** 2, label = "exact")
plt.plot(x, np.abs(evecs[:,n_plot])**2 / dx, label = "numerical")
plt.title("exact and numerical")
plt.xlabel("x")
plt.ylabel("$|\\phi_n(x)|^2$")
plt.legend()

# plot relative error
plt.figure()
plt.plot(x, np.abs(np.abs(cqd.hamiltonians.HO_eigenstates_exact(n_plot, x)) ** 2 - np.abs(evecs[:, n_plot]) ** 2 / dx))
plt.title("absolute probability error")
plt.xlabel("x")
plt.ylabel("error");
```

Observations:

If the grid is too small, the ground state wave function gets narrower due to the hard wall boundary conditions. All the eigenenergies lie above the exact ones.

If too few gridpoints are used, the curvature of the wave function at the center is not well resolved. The numerically obtained energies are below the exact ones, probably because the kinetic term is underestimated.
