# Computational Complexity across Disciplines

## 1. General Complexity Theory
In general computer science, the two primary computational resources are:
* **Time** — how many computational steps are required.
* **Space (Memory)** — how much memory is required.

These are the resources used to define the most famous complexity classes:
* **$\text{P}$**: Polynomial time
* **$\text{NP}$**: Nondeterministic polynomial time
* **$\text{PSPACE}$**: Polynomial memory
* **$\text{EXPTIME}$**: Exponential time

## 2. Classical Computing (CC)
Traditionally, complexity theory focuses heavily on:
$$\boxed{\text{Time Complexity}} \quad \text{and} \quad \boxed{\text{Space Complexity}}$$
Because they are machine-independent and mathematically tractable.

### Examples:
* **Sorting:** $\mathcal{O}(N \log N)$ time
* **Matrix multiplication:** $\mathcal{O}(N^3)$ (or better)
* **Sparse matrix-vector multiplication:** $\mathcal{O}(\text{nnz})$

---

## 3. Quantum Computing (QC)
In quantum complexity theory, we still care about time and memory, but they take on quantum analogs:

### Quantum Time
Usually measured by:
* Number of quantum gates
* Circuit depth
*(Instead of standard CPU instructions)*

### Quantum Memory
Measured by:
* Number of qubits

> **Example:** Shor's algorithm uses a polynomial gate count and a polynomial number of qubits. A quantum simulation may require $n$ qubits to represent an $n$-particle system.

---

## 4. Computational Physics (CP)
In computational physics, theoretical bounds translate into very practical, physical limitations:
* CPU time
* RAM
* Communication cost (on parallel clusters/supercomputers)

> **Example:** When solving a sparse eigenvalue problem:
> * **Time** = number of matrix-vector multiplications performed.
> * **Memory** = storage required for the Hamiltonian matrix and state vectors.

---

## 5. The Role of Energy
While energy consumption is increasingly vital in physical engineering, it is not usually a primary resource in standard complexity theory. Instead, researchers study it under specialized fields like:
* Energy complexity
* Thermodynamics of computation
* Reversible computing / Quantum thermodynamics

In most computer science and quantum courses, complexity classes are not defined using energy because energy is often treated as a consequence of hardware activity:
$$\text{Energy} \propto \text{operations performed} + \text{memory accesses}$$

If you are studying numerical methods, sparse matrices, Krylov subspace methods, quantum simulation, or computational physics, the two resources discussed most often will always be:

# Sparse Matrices in Computational Physics

When working with a sparse Hamiltonian $H$, we primarily care about two types of resource complexities:

## 1. Memory Resource
> **Key Question:** Can $H$ fit into RAM?

* **Dense Representation:** $\mathcal{O}(N^2)$ memory scaling.
* **Sparse Representation:** $\mathcal{O}(\text{nnz})$ memory scaling, where $\text{nnz}$ is the number of non-zero elements.


## 2. Time Resource
> **Key Question:** How many operations are required to compute the matrix-vector product $H|\psi\rangle$?

* **Dense Representation:** $\mathcal{O}(N^2)$ operations.
* **Sparse Representation:** $\mathcal{O}(\text{nnz})$ operations.


## Summary of Terms

When physics lectures or textbooks refer to resource complexities, they mean:

* **Time Complexity:** $$\boxed{\text{Number of computational steps (operations)}}$$
* **Memory Complexity:** $$\boxed{\text{Amount of storage (RAM) required}}$$

> 💡 **Note on Runtime:** The actual CPU runtime is closely related to time complexity, but complexity theory specifically measures the *growth* of the operation count as the system size $N$ becomes large.

## `np.linalg.eigh` vs `scipy.sparse.linalg.eigsh`

Both functions are used to compute:
- eigenvalues
- eigenvectors

```python
eigenvalues, eigenvectors = np.linalg.eigh(H)

eigenvalues, eigenvectors = eigsh(H, k=2)
```

- first returned object → eigenvalues
- second returned object → eigenvectors

### Main difference

| Function | Best for |
|---|---|
| `np.linalg.eigh` | small dense matrices |
| `eigsh` | large sparse matrices |

### Why `eigsh` is important

In quantum computing and numerical simulations, memory is very important.

Large Hamiltonians can become huge, so:

```python
np.linalg.eigh(H)
```

may become:
- slow
- memory expensive

`eigsh` is more efficient because it computes only a few important eigenvalues/eigenvectors, such as:
- ground state energy
- lowest energy states

So `eigsh` is better for large quantum systems.

### Why use `np.linalg.eigh`?

`np.linalg.eigh` is useful for:
- small systems
- dense matrices
- computing all eigenvalues and eigenvectors exactly

### Simple idea

| Function | Best use |
|---|---|
| `eigh` | exact full diagonalization |
| `eigsh` | efficient for large sparse systems |

# Sparse Matrix Formats in SciPy

## 1. COO Format (`"coo"`)

**COO = Coordinate Format**

Stores only non-zero elements.

### Best for:
- Building sparse matrices
- Adding elements easily

## 2. CSR Format (`"csr"`)

**CSR = Compressed Sparse Row**

Stores non-zero elements row-by-row.

### Best for:
- Fast matrix computations
- Matrix-vector multiplication
- Quantum Hamiltonians

### Advantage:
Efficient for numerical simulations.

## 3. CSC Format (`"csc"`)

**CSC = Compressed Sparse Column**

Stores non-zero elements column-by-column.

### Best for:
- Column operations
- Sparse matrix factorizations

### Advantage:
Efficient for column slicing.


## 4. DIA Format (`"dia"`)

**DIA = Diagonal Format**

Stores only the diagonals of a matrix.

### Best for:
- Tridiagonal matrices
- Finite difference Hamiltonian

Only the diagonals are stored.
